```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.api import acf, pacf, ARIMA, ArmaProcess, AutoReg, adfuller
from statsmodels.tsa import arima_process
```

```python
np.random.seed(100)
x_AR1 = arima_process.arma_generate_sample(ar=[1, 0.5], ma=[1], nsample=200)
np.random.seed(100)
x_ar1 = ArmaProcess(ar=[1, 0.5]).generate_sample(nsample=200)
```

```python
plt.plot(x_ar1)
plt.show()
```

```python
plt.plot(x_AR1)
plt.show()
```

```python
plot_acf(x_AR1)
plt.show()
```

```python
plot_pacf(x_AR1)
plt.show()
```

```python
model = ARIMA(x_AR1, order=(1, 0, 0)).fit()
model.summary()
```
